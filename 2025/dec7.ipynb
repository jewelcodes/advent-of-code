# Dec 7 - Part 1

```python
raw = open("inputs/dec7.txt").read()
splitters = [(x, y) for x in range(len(raw.split("\n")[0]))
             for y in range(len(raw.split("\n"))) if raw.split("\n")[y][x] == "^"]
beams = set([(x, y+1) for x in range(len(raw.split("\n")[0]))
             for y in range(len(raw.split("\n"))) if raw.split("\n")[y][x] == "S"])
og_beams = beams
```

```python
count = 0
while all(y < len(raw.splitlines()) for _, y in beams):
    beams = set([(x, y+1) for x, y in beams])
    splits = beams.intersection(splitters)
    beams -= splits
    beams.update([(x+dx, y) for x, y in splits for dx in (1, -1)
                  if x+dx >= 0 and x+dx < len(raw.splitlines()[0])])
    count += len(splits)

print(count)
```

# Dec 7 - Part 2

```python
cache = {}
def travel(x, y):
    if x < 0 or x >= len(raw.splitlines()[0]) or y >= len(raw.splitlines()):
        return 0

    if (x, y+1) in splitters:
        l, r = (x-1, y+1), (x+1, y+1)
        if l not in cache:
            cache[l] = travel(*l)
        if r not in cache:
            cache[r] = travel(*r)
        return 1 + cache[l] + cache[r]

    return travel(x, y+1)
```

```python
count = travel(*og_beams.pop()) + 1
print(count)
```
