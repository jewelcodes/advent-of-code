# Dec 9 - Part 1

```python
raw = open("inputs/dec9.txt").read()
tiles = [tuple([int(c) for c in l.split(",")]) for l in raw.splitlines()]
```

```python
def area(x1, y1, x2, y2):
    return (abs(x2-x1) + 1) * (abs(y2-y1) + 1)

areas = {(c1, c2): area(*c1, *c2) for c1 in tiles for c2 in tiles if c1 < c2}
print(max(areas.values()))
```

# Dec 9 - Part 2

```python
outline = [(s, e) for s, e in zip(tiles + [tiles[0]], (tiles + [tiles[0]])[1:])]
```

```python
def is_inside(x1, y1, x2, y2):
    xs, xe = min(x1, x2), max(x1, x2)
    ys, ye = min(y1, y2), max(y1, y2)

    for s, e in outline:
        l_xs, l_xe = min(s[0], e[0]), max(s[0], e[0])
        l_ys, l_ye = min(s[1], e[1]), max(s[1], e[1])
        if xs < l_xe and xe > l_xs and ys < l_ye and ye > l_ys:
            return False
    return True
```

```python
areas = {(c1, c2): area(*c1, *c2) for c1 in tiles for c2 in tiles
         if c1 < c2 and is_inside(*c1, *c2)}
print(max(areas.values()))
```
