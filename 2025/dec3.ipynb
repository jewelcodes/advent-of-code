# Dec 3 - Part 1

```python
with open("inputs/dec3.txt") as f:
    lines = f.readlines()
```

```python
total = 0
for l in lines:
    d1 = max(l[:len(l) - 2])
    d2 = max(l[l.index(d1) + 1:])
    total += int(d1) * 10 + int(d2)

print(total)
```

# Dec 3 - Part 2

```python
total = 0
for l in lines:
    start, end = 0, len(l) - 12
    for i in range(12):
        total += int(max(l[start:end])) * (10 ** (11 - i))
        start += l[start:end].index(max(l[start:end])) + 1
        end += 1

print(total)
```
