# Dec 11 - Part 1

```python
raw = open("inputs/dec11.txt").read()
devices = {l.split(":")[0]: [c for c in l.split(":")[1].split(" ") if c]
           for l in raw.splitlines()}
```

```python
def count_paths(query):
    outputs = devices[query]
    count = 0
    for output in outputs:
        count += 1 if output == "out" else count_paths(output)
    return count

print(count_paths("you"))
```

# Dec 11 - Part 2

```python
from functools import lru_cache

@lru_cache(maxsize=1024)
def count_paths(query, fft=False, dac=False):
    fft |= query == "fft"
    dac |= query == "dac"

    outputs = devices[query]
    count = 0
    for output in outputs:
        if output != "out":
            count += count_paths(output, fft, dac)
        elif dac and fft:
            count += 1
    return count

print(count_paths("svr"))
```
