# Dec 5 - Part 1

```python
raw = open("inputs/dec5.txt").read()
fresh = [tuple([int(x) for x in r.split("-")])
         for r in raw.split("\n\n")[0].split("\n")]
available = [int(x) for x in raw.split("\n\n")[1].split("\n")]
```

```python
count = sum(any(f[0] <= i <= f[1] for f in fresh) for i in available)
print(count)
```

# Dec 5 - Part 2

```python
sorted_fresh = sorted(fresh, key=lambda f: f[0])
merged = []
for x, y in sorted_fresh:
    if not merged or x > merged[-1][1]:
        merged.append((x, y))
    else:
        merged[-1] = (merged[-1][0], max(merged[-1][1], y))

count = sum(f[1] - f[0] + 1 for f in merged)
print(count)
```
