# Dec 1 - Part 1

```python
with open("inputs/dec1.txt") as f:
    lines = f.readlines()
```

```python
rotations = [int(l[1:]) * -1 if l[0] == 'L' else int(l[1:]) for l in lines]
current, zeroes = 50, 0

for r in rotations:
    current = (current + r) % 100
    if not current:
        zeroes += 1

print(zeroes)
```

# Dec 1 - Part 2

```python
current, zeroes = 50, 0

for r in rotations:
    temp = current + r
    if temp < 0:
        zeroes += (0 if not current else 1) + (abs(temp) // 100)
    elif temp > 0:
        zeroes += temp // 100
    else:
        zeroes += 1 + (1 if not current else 0)

    current = (current + r) % 100

print(zeroes)
```
