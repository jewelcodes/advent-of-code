# Dec 8 - Part 1

```python
raw = open("inputs/dec8.txt").read()
junctions = [tuple([int(c) for c in l.split(",")]) for l in raw.split("\n")]
```

```python
def d(x1, y1, z1, x2, y2, z2):
    import math
    return math.sqrt(((x2-x1)**2) + ((y2-y1)**2) + ((z2-z1)**2))

distances = {(j1, j2): d(*j1, *j2) for j1 in junctions
             for j2 in junctions if j1 < j2}
```

```python
nearest_pairs = sorted(distances.keys(), key=lambda k: distances[k])[:1000]
circuits = []
for j1, j2 in nearest_pairs:
    if not any(j1 in c or j2 in c for c in circuits):
        circuits.append(set([j1, j2]))
    for c in circuits:
        if j1 in c or j2 in c:
            c.update([j1, j2])

    dups = [i for i, c in enumerate(circuits) if j1 in c or j2 in c]
    for i in dups[1:]:
        circuits[dups[0]] = circuits[dups[0]].union(circuits[i])
    for i in dups[1:]:
        del circuits[i]

count = 1
for l in sorted([len(c) for c in circuits], reverse=True)[:3]:
    count *= l
print(count)
```

# Dec 8 - Part 2

```python
nearest_pairs = sorted(distances.keys(), key=lambda k: distances[k])
unconnected, last_connection = junctions.copy(), None
for j1, j2 in nearest_pairs:
    if j1 in unconnected:
        del unconnected[unconnected.index(j1)]
    if j2 in unconnected:
        del unconnected[unconnected.index(j2)]
    if not unconnected:
        last_connection = (j1, j2)
        break

print(last_connection[0][0] * last_connection[1][0])
```
