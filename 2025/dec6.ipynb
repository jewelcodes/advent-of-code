# Dec 6 - Part 1

```python
raw = open("inputs/dec6.txt").read()
data = [[s.strip() for s in r.split()] for r in raw.split("\n")]
data = [[data[r][c] for r in range(len(data))] for c in range(len(data[0]))]
```

```python
total = 0
for col in data:
    if col[-1] == "+":
        total += sum(int(c) for c in col[:-1])
    else:
        prod = 1
        for n in col[:-1]:
            prod *= int(n)
        total += prod

print(total)
```

# Dec 6 - Part 2

```python
lines = raw.splitlines()
digit_lines, ops_line = lines[:-1], lines[-1]
cols_with_ops = [i for i, ch in enumerate(ops_line) if ch in ("*", "+")]
widths = [cols_with_ops[i+1] - cols_with_ops[i] \
          for i in range(len(cols_with_ops) - 1)]

data2 = [
    [
        "".join(r[i] for r in digit_lines if r[i].isdigit())
        for i in range(min(j + min(widths), len(ops_line)) - 1, j - 1, -1)
        if any(r[i].isdigit() for r in digit_lines)
    ] + [ops_line[j]]
    for j in cols_with_ops
]

total = 0
for col in data2:
    if col[-1] == "+":
        total += sum(int(c) for c in col[:-1])
    else:
        prod = 1
        for n in col[:-1]:
            prod *= int(n)
        total += prod

print(total)
```
