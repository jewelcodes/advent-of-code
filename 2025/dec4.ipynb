# Dec 4 - Part 1

```python
with open("inputs/dec4.txt") as f:
    lines = [list(l.strip()) for l in f.readlines()]
```

```python
total = 0
for y in range(len(lines)):
    for x in range(len(lines[0])):
        if lines[y][x] != "@":
            continue
        coords = [(x+dx, y+dy) for dx in (-1, 0, 1) for dy in (-1, 0, 1) \
                  if (not dx == dy == 0) and 0 <= (x+dx) < len(lines[0]) \
                  and 0 <= (y+dy) < len(lines)]
        if sum(lines[dy][dx] == "@" for dx, dy in coords) < 4:
            total += 1

print(total)
```

# Dec 4 - Part 2

```python
total, subtotal = 0, float("inf")
while subtotal:
    subtotal = 0
    for y in range(len(lines)):
        for x in range(len(lines[0])):
            if lines[y][x] != "@":
                continue
            coords = [(x+dx, y+dy) for dx in (-1, 0, 1) for dy in (-1, 0, 1) \
                      if (not dx == dy == 0) and 0 <= (x+dx) < len(lines[0]) \
                      and 0 <= (y+dy) < len(lines)]
            if sum(lines[dy][dx] == "@" for dx, dy in coords) < 4:
                lines[y][x] = "."
                subtotal += 1

    total += subtotal

print(total)
```
