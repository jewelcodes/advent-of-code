# Dec 2 - Part 1

```python
with open("inputs/dec2.txt") as f:
    raw = f.read()
```

```python
def repeated_digits(id_range):
    repeats = []
    start, end = id_range
    digits = [i // 2 for i in range(len(str(start)), len(str(end))+1) \
              if not i % 2]
    
    for i in digits:
        repeat_sequence = str(start)[:i] if len(str(start)) \
            == len(str((end))) else "1" + "".join(["0"] * (i - 1))

        while len(repeat_sequence) == i:
            number = int(repeat_sequence + repeat_sequence)
            if start <= number <= end:
                repeats.append(number)
            elif number > end:
                break

            repeat_sequence = str(int(repeat_sequence) + 1)

    return repeats
```

```python
ranges = [tuple([int(x) for x in l.split("-")]) for l in raw.split(",")]
password = 0
for r in ranges:
    password += sum(repeated_digits(r))

print(password)
```

# Dec 2 - Part 2

```python
def repeated_digits2(id_range):
    repeats = []
    start, end = id_range
    digits = [i for n in (start, end) for i in range(1, len(str(n))) \
              if not len(str(n)) % i]

    for i in digits:
        multiples = []
        if len(str(start)) != 1 and not len(str(start)) % i:
            multiples.append(len(str(start)) // i)
        if not len(str(end)) % i:
            multiples.append(len(str(end)) // i)

        for m in multiples:
            repeat_sequence = str(start)[:i] if len(str(start)) \
                == len(str((end))) else "1" + "".join(["0"] * (i - 1))
            while len(repeat_sequence) == i:
                number = int(repeat_sequence * m)
                repeat_sequence = str(int(repeat_sequence) + 1)

                if number not in repeats and start <= number <= end:
                    repeats.append(number)
                elif number > end:
                    break

    return repeats
```

```python
password = 0
for r in ranges:
    password += sum(repeated_digits2(r))

print(password)
```
